```python
from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/Fiap/'
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("AmazonDataSpark").getOrCreate()
```

```python
# Carregar arquivos formato parquet para um dataframe pyspark
df = spark.read.parquet(file_path + 'database_limpo')
df.createOrReplaceTempView('df')
df.count()
```

```python
# Pegar 100000 registros randomicos do df
sample_df = df.sample(withReplacement=False, fraction=0.1, seed=42).limit(100000)
sample_df = sample_df.withColumnRenamed("title_clear", "title").withColumnRenamed("content_clear", "content").toPandas()
```
